```python
import pandas as pd
import numpy as np 

df = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv", encoding = "Latin-1")
#df_e = df.groupby("Element")['Stock Variation'].max()
df_e = df.groupby("Element")
for group_name, group_data in df_e:
    print(f'Group name: {group_name}')
    print(group_data)
df_e.max()
```

```python
#finding the country with the highest sum in 2017
df_y = df[['Y2017','Area']].groupby('Area')
for x,y in df_y:
    print(f'Group_name: {x}')
    print(y['Y2017'].sum())
```

```python
#year with the highest sum of stock variation
df_y = df.groupby('Element')
for x,y in df_y:
    if x == 'Stock Variation':
           print(y[['Y2014', 'Y2015', 'Y2016', 'Y2017', 'Y2018']].sum())
```

```python
#Total number and percentage of missing data in 2014
n_sum = df['Y2014'].isnull().sum()
per = (int(n_sum)/len(df['Y2014']))* 100
print(f'Sum = {n_sum}\nPercentage = {round(per, 3)}')
```

```python
#The sum of wine produced in 2015 and 2018
df_i = df.groupby("Item")
for x,y in df_i:
    if x == 'Wine':
        print(y[['Y2015','Y2018']].sum())
```

```python
#printing the total number of unique countries 
df_u = df['Area'].unique()
print(len(df_u))
```

```python
#sum of processing in Y2017
df_y = df.groupby('Element')
for x,y in df_y:
    if x == 'Processing':
        print(y['Y2017'].sum())
```
